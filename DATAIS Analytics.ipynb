```python
df = df.loc[df.pdays != -1]
df
```

```python
df.age.max()
```

```python
def age_(data):  
    if(len(str(data))) == 1:
        data = 0 
    else:
        data= int(str(data)[0])
    
    return data
            
```

```python
df["range_age"] = df['age'].apply(age_)
```

```python
def transform_bool(data):
    
    if data == "yes":
        data = True
    elif data == 'no':
        data = False
    else:
        pass
    return data
```

```python
for i in df.columns:
    df[i]= df[i].apply(transform_bool)
```

```python
df[['month','day']]
```

```python
def change_month(data):
    from time import strptime
    
    month = strptime(data,'%b').tm_mon
    return month

    
```

```python
df['Month_number'] = df['month'].apply(change_month)
```

```python
df["outcome"]= df["Month_number"].map(str) + "/" + df["day"].map(str)
```
